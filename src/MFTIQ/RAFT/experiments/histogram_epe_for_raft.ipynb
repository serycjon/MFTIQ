```python
import os.path

import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
save_root = '/datagrid/personal/neoral/RAFT_occl_uncertainty_outputs/base-sintel/statistics_graphs'

save_path1 = os.path.join(save_root, 'sintel_clean_training.txt')
save_path2 = os.path.join(save_root, 'sintel_final_training.txt')
```

```python
hist1 = np.loadtxt(save_path1)
hist2 = np.loadtxt(save_path2)
hist_both = hist1+hist2
nbins = 50
thresholds = np.array(list(range(nbins+1))) + 0.5
```

```python
plt.bar(thresholds, hist1)
plt.yscale('log')
```

```python
hist_norm = hist_both[:-1] / np.sum(hist_both[:-1])
```

```python
print(hist_norm.shape)
print(thresholds.shape)
print(thresholds)
```

```python
pol_res = np.polyfit(thresholds[:-1], np.log(hist_norm), 10)
print(pol_res)

p = np.poly1d(pol_res)
y = p(thresholds[:-1])
print(y)

plt.figure()
plt.bar(thresholds[:-1], hist_norm)
plt.plot(thresholds[:-1], np.exp(y), 'r')
plt.yscale('log')
plt.show()
plt.close()

plt.figure()
plt.bar(thresholds[:-1], hist_norm)
plt.plot(thresholds[:-1], np.exp(y), 'r')
# plt.yscale('log')
plt.show()
plt.close()

print(np.sum(np.exp(y)))
```

```python
pol_res = np.polyfit(thresholds[:-1], 1/hist_norm, 3)
print(pol_res)

p = np.poly1d(pol_res)
y = p(thresholds[:-1])
print(y)

plt.figure()
plt.bar(thresholds[:-1], 1/hist_norm)
plt.plot(thresholds[:-1], y, 'r')
plt.yscale('log')
plt.show()
plt.close()

plt.figure()
plt.bar(thresholds[:-1], 1/hist_norm)
plt.plot(thresholds[:-1], y, 'r')
# plt.yscale('log')
plt.show()
plt.close()

print(np.sum(np.exp(y)))
```
